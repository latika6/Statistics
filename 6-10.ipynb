```python
#6 Perform a chi-square test of independence for two categorical variables
import scipy.stats as stats

# Create a contingency table
observed = [[10, 20], [30, 40]]  # Example values

# Perform chi-square test
chi2, p, _, _ = stats.chi2_contingency(observed)

# Print the results
print("Chi-square Statistic:", chi2)
print("p-value:", p)
```

```python
#7 Calculate the probability of a given event occurring in a binomial distribution
from scipy.stats import binom

# Number of trials
n = 5

# Probability of success (getting a head)
p = 0.5

# Number of successes
k = 3

# Calculate the probability using the PMF formula
probability = binom.pmf(k, n, p)

print("Probability of getting exactly 3 heads in 5 coin flips:", probability)
```

```python
#8. Fit a linear regression model to a dataset and interpret the coefficients
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample dataset
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable (feature)
y = np.array([2, 4, 5, 4, 5])  # Dependent variable (target)

# Fit the linear regression model
model = LinearRegression().fit(X, y)

# Get the coefficients
intercept = model.intercept_
slope = model.coef_[0]

print("Intercept:", intercept)
print("Slope:", slope)
```

```python
#9. Create a histogram of a dataset and determine the appropriate number of bins
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

# Create a histogram with different bin sizes
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title('Histogram with 5 bins')

plt.subplot(2, 2, 2)
plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram with 10 bins')

plt.subplot(2, 2, 3)
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram with 20 bins')

plt.subplot(2, 2, 4)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram with 30 bins')

plt.tight_layout()
plt.show()
```

```python
# 10. Calculate the confidence interval for a population mean
import numpy as np
from scipy import stats

# Sample data
data = np.array([10, 12, 15, 14, 16, 18, 11, 13, 14, 16])

# Sample mean
sample_mean = np.mean(data)

# Population standard deviation (assuming known for this example)
population_stddev = 2.5

# Z-score for a 95% confidence interval
confidence_level = 0.95
Z = stats.norm.ppf((1 + confidence_level) / 2)  # 1.96 for a 95% confidence interval

# Sample size
n = len(data)

# Calculate the confidence interval
margin_of_error = Z * (population_stddev / np.sqrt(n))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("Sample Mean:", sample_mean)
print("Confidence Interval:", confidence_interval)
```
